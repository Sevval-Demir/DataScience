```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**2
```

```python
numpyDizisi2
```

```python
(benimFigur,benimEksen)=plt.subplots()
```

```python
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
benimEksen.plot(numpyDizisi2,numpyDizisi1,"b")
```

```python
(yeniFigur,yeniEksen)=plt.subplots()
```

```python
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,color="blue",linewidth=4.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,color="yellow")
plt.show()
```

## scatter

```python
plt.scatter(numpyDizisi1,numpyDizisi2)
plt.show()
```
