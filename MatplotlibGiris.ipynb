```python
myFigure=plt.figure()
figureAxes=myFigure.add_axes([0.4,0.4,0.4,0.4])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri smi")
figureAxes.set_ylabel("Y Ekseni Veri smi")
figureAxes.set_title("Grafik Bal")
plt.show()
```

```python
figure2=plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2=figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Balk")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Kk Grafik Balk")
plt.show()
```
